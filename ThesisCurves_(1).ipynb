```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for plotting
data = {
    'n1': [1, 2, 3, 4, 5, 6, 7],
    'Omega1': [0.9323, 0.7748, 0.7109, 0.7924, 0.9959, 1.26, 1.589],
    'n2': [1, 2, 3, 4, 5, 6, 7],
    'Omega2': [1.046, 0.9698, 1.053, 1.3216, 1.6979, 2.1505, 2.6554],
    'Omega3': [1.1823, 1.1638, 1.3299, 1.7086, 2.1694, 2.6967, 3.2661]
}

# Create DataFrame
df_manual = pd.DataFrame(data)

# Plot all three datasets
plt.figure(figsize=(8, 6))
plt.plot(df_manual['n1'], df_manual['Omega1'], label="h/r=.1", marker='o')
plt.plot(df_manual['n2'], df_manual['Omega2'], label="h/r=.2", marker='x')
plt.plot(df_manual['n1'], df_manual['Omega3'], label="h/r=.3", marker='s')

plt.title('Dimensionless frequency vs wave numbers')
plt.xlabel('n')
plt.ylabel(r'$\Omega = \omega r \sqrt{\frac{\rho}{E}}$')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
